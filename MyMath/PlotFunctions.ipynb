## Plotten; Symbolische Aktionen; Zwei Ableitungen

```python
%config InlineBackend.figure_format = 'retina' # erhöht die Auflösung der plots
from sympy import *
init_printing()
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from IPython.display import display, Math
```

```python
def ableitungen(term):
    term_d = diff (term, x)
    term_dd = diff (term_d, x)    
    return term_d, term_dd

def functions(term):
    f = Lambda(x,term)(x)
    term_d, term_dd = ableitungen(term)
    g = Lambda(x,term_d)(x)
    h = Lambda(x,term_dd)(x)
    return f, g, h, term_d, term_dd
    
def plotFkt(term, ugx, ogx, ugy, ogy):
    f, g, h, term_d, term_dd = functions(term)
    p1=plot(f,(x,ugx,ogx),show=False)
    p2=plot(g,(x,ugx,ogx),show=False)
    p3=plot(h,(x,ugx,ogx),show=False)
    p1.extend(p2)
    p1.extend(p3)
    p1.legend=True
    
    fl = "$" + "f(x) = " + latex(term) + "$"
    
    fdl = "$" + "f^{\prime}(x) = " + latex(term_d) + "$"

    fddl = "$" + "f^{\prime\prime}(x) = " + latex(term_dd) + "$"
    
    p1[0].label = fl
    p1[1].label = fdl
    p1[2].label = fddl
    
    p1[1].line_color='red'
    p1[2].line_color='green'
    
    p1.ylim=(ugy, ogy)
    p1.show() 
    
    display (Math (latex(fl)))
    display (Math (latex(fdl)))
    display (Math (latex(fddl)))
```
