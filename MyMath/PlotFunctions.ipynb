## Plotten; Symbolische Aktionen; Zwei Ableitungen und Stammfunktion

```python
%config InlineBackend.figure_format = 'retina' # erhöht die Auflösung der plots
from sympy import *
init_printing()
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from IPython.display import display, Math, Latex
```

```python
def ableitungen(term):
    term_d = diff (term, x)
    term_dd = diff (term_d, x)    
    return term_d, term_dd

def functions(term):
    f = Lambda(x,term)(x)
    term_d, term_dd = ableitungen(term)
    g = Lambda(x,term_d)(x)
    h = Lambda(x,term_dd)(x)
    return f, g, h, term_d, term_dd
    
def plotFkt(term, ugx, ogx, ugy, ogy):
    f, g, h, term_d, term_dd = functions(term)
    p1=plot(f,(x,ugx,ogx),show=False)
    p2=plot(g,(x,ugx,ogx),show=False)
    p3=plot(h,(x,ugx,ogx),show=False)
    p1.extend(p2)
    p1.extend(p3)
    p1.legend=True
    
    fl = "$" + "f(x) = " + latex(term) + "$"
    
    fdl = "$" + "f^{\prime}(x) = " + latex(term_d) + "$"

    fddl = "$" + "f^{\prime\prime}(x) = " + latex(term_dd) + "$"
    
    p1[0].label = fl
    p1[1].label = fdl
    p1[2].label = fddl
    
    p1[1].line_color='red'
    p1[2].line_color='green'
    
    p1.ylim=(ugy, ogy)
    p1.show() 
    
    display (Math (fl))
    display (Math (fdl))
    display (Math (fddl))
```

```python
def stamm (term):
    return integrate(term,x)
```

```python
def aufleit(term, verschiebung):
    f = Lambda(x,term)(x)
    term_stamm = stamm(term) + verschiebung
    f_stamm = Lambda(x,term_stamm)(x)
    return f, f_stamm, term_stamm
```

```python
def plotAufleitFkt(term, ugx, ogx, ugy, ogy, verschiebung):
    f, f_stamm, term_stamm = aufleit(term, verschiebung)
    p1=plot(f,(x,ugx,ogx),show=False)
    p2=plot(f_stamm,(x,ugx,ogx),show=False)
    p1.extend(p2)
    p1.legend=True
    
    fl = "$" + "f(x) = " + latex(term) + "$"
    
    fstamml = "$" + "F(x) = " + latex(term_stamm) + "$"
    
    p1[0].label = fl
    p1[1].label = fstamml
    
    p1[1].line_color='red'
    
    p1.ylim=(ugy, ogy)
    p1.show() 
        
    display (Math (fl))
    display (Math (fstamml))
```

```python
x=Symbol('x')
#plotAufleitFkt (-(x-2)*(x+2)/3, -4, 4, -4, 2, -2)
#plotAufleitFkt ((-8)*(x**3)/27 + 2*x - 2, -4, 5, -5, 15, 0)
plotAufleitFkt (x*sin(x), -4, 5, -5, 15, 0)
```

Wieso ist die Stammfunktion (rot) nicht (achsen)symmetrisch?

Eine mögliche Argumentation:

- Die Funktion (blau) hat an dert Stelle 1.5 einen Extrempunkt und eine waagerechte Tangente. Also hat die Stammfunktion dort einen Sattelpunkt. Das ist also sozusagen eine doppelte Extremstelle.
- Die Stammfunktion hat ein Extremum an der Stelle -3.
- Bei einer Funktion 4-ten Grades kann es nur höchstens 3 Extrema geben, so dass es keinen weiteren Extrempunkt geben kann.
