```python
# Datensatz einlesen
# Die Datei "Käfer.csv" ist eine Datei, in der Datensätze zeilenweise aufgeführt sind.
# Jede Spalte ist durch ein Semikolon (";") getrennt.

# df_käfer ist eine eigene Variable und steht für "dataframe Käfer"
df_käfer = pd.read_csv('Käfer.csv', sep=';')

# zeige den Datensatz "df_käfer" an
display(df_käfer)
```

## Wie kann ich den gesamten Datensatz ausgeben?

```python
# Einstellung zur Anzeige aller Datensätze (Zeilen)
pd.set_option('display.max_rows', None)
# Einstellung zur Anzeige aller Spalten
pd.set_option('display.max_columns', None)

# Datensatz ausgeben
display(df_käfer)
```

## Wie kann ich den Datensatz sortieren?

```python
# sortiere die Datensätze nach einer Spalte (aufsteigend) mit Ausgabe
sortInsekt = df_käfer.sort_values(by='Insekt')
display(sortInsekt)
```

# Vertiefung für die Schnellen

Im Folgenden wird das Streudiagramm mit den 100 Marienkäfern und 100 Raupen dargestellt.
Bitte folgende Code-Zellen ausführen.

```python
# Dieser Quelltext ist nicht prüfungsrelevant.

#Bibliotheken plotly und cufflinks zur Datenvisualisierung importieren
from plotly.offline import iplot
import plotly.graph_objects as go
import cufflinks as cf
cf.go_offline()


#Ein quadratsiches Layout für plotly-Graphiken wird definiert 
square_layout = go.Layout(xaxis={'title':'Breite','range':[-0.01,1.01]},
                   yaxis={'title':'Länge','range':[-0.01,1.01]},
                   height=810,
                   width=810)
```

```python
# Dieser Quelltext ist nicht prüfungsrelevant.

df_käfer.iplot('scatter', mode = 'markers',  layout = square_layout,
               x = 'Breite', 
               y = 'Länge',
               categories = 'Insekt')
```
