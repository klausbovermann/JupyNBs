## Aktivierungsfunktionen

```python
# scipy.special for the sigmoid function expit()
import scipy.special
import plotly.graph_objects as go
import pandas as pd
from plotly.offline import iplot
import cufflinks as cf
import numpy as np
cf.go_offline()
```

```python
def aktivierung_sigmoid_1(x):
    #return scipy.special.expit(x)   # intern imp
    return np.exp(x)/(1+np.exp(x))

def aktivierung_tanh(x):
    return np.tanh(x)
```

```python
#Den Code in dieser Zelle musst du nicht verstehen - Wichtig ist die erzeugt Visualisierung
#Layout der Graphik festlegen
aktivierung_layout = go.Layout(xaxis={'title':'Input','range':[-10.0,10.0]},
                               yaxis={'title':'Aktivierung','range':[-1.5,1.5]},
                               height=600,
                               width=800)

#Wertetabelle erzeugen
df_aktivierung = pd.DataFrame()
df_aktivierung['Input']=[x/100 for x in range(-1000, 1000)]
df_aktivierung['sigmoid_intern']=[aktivierung_sigmoid_1(x/100) for x in range(-1000, 1000)]
df_aktivierung['tanh']=[aktivierung_tanh(x/100) for x in range(-1000, 1000)]

#Werte als Graphik Visualisieren
df_aktivierung.iplot(x='Input', y=['tanh','sigmoid_intern'] , layout = aktivierung_layout)
```
